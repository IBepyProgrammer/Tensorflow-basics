```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("churn.csv")
```

```python
df
```

```python
# The get_dummies converts categorical data into numerical dummy data
# ( eg splits the gender columns into male and female columns)etc(converts yes ,no to 1's and 0's)

X = pd.get_dummies(df.drop(["Churn", "Customer ID"], axis = 1))
y = df["Churn"].apply(lambda x: 1 if x=="yes" else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2)
```

# We then import our dependancies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

# We then build the model

```python
model = Sequential()
model.add(Dense(units = 32, activation = "relu", input_dim = len(X_train.columns)))
model.add(Dense(units = 64, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "sgd", metrics = "accuracy")
```

# We then fit our model

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```

# We can the save the model and then Reload from memory

```python
model.save("TFSample_model")
```

```python
# this deletes the model from memory in out jupyter notebook

del model
```
